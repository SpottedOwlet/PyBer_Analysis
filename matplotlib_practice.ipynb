```python
%matplotlib inline
# Import dependencies.
import matplotlib.pyplot as plt
```

### Line Plots
#### MATLAB Syntax Method:

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# create the plot using Matlab's plotting syntax
plt.plot(x_axis,y_axis)
```

### Line Plot
#### Object-Oriented Interface Method: 

```python
# Create the plot with ax.plt() an object-oriented interface method
figure, axes = plt.subplots(figsize=(6,4))
axes.plot(x_axis,y_axis)
```

```python
# alternatively above code can be written as:
# create plot with ax.plt
fig = plt.figure()
ax = fig.add_subplot()
ax.plot(x_axis,y_axis)
```

```python
# Create the plot with ax.plt(): to create a quick graph you can use following two lines
ax = plt.axes()
ax.plot(x_axis,y_axis)
```

### Use plt.show()


```python
# Create the plot.
plt.plot(x_axis, y_axis)
plt.show()
```

### Annotate The Chart using MATLAB syntax

```python
# Create the plot and add a label for the legend.
plt.plot(x_axis,y_axis,label = 'Boston')
# create labels for x and y axes.
plt.xlabel('Date')
plt.ylabel("Fare($)")
# set the y limit between 0 and 45
plt.ylim(0, 45)
# create a title
plt.title("PyBer Fare by Month")
# add the legend
plt.legend()
```

```python
# Create the plot.
plt.plot(x_axis,y_axis,marker = '*',color = "blue",linewidth =2,label = 'Boston')
# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Set the y limit between 0 and 45.
plt.ylim(0, 45)
# Create a title.
plt.title("PyBer Fare by Month")
# Add a grid.
plt.grid()
# Add the legend.
plt.legend()
```

### Annotate The Chart using Object-Oriented method

```python
# create the plot
fig,ax = plt.subplots(figsize = (6,4))
# add a label
ax.plot(x_axis,y_axis, marker = 'D', color = "green",linewidth = 2, label = 'Boston')
# add axes names
ax.set_xlabel('Date')
ax.set_ylabel('fare($)')
# add title
ax.set_title('PyBer fare by Month')
# add y-axis limit
ax.set_ylim(0,45)
ax.set_xlim("Jan","Dec")
# add grid
ax.grid()
# add legend
ax.legend()
```

```python
# Import dependencies.
import matplotlib.pyplot as plt
import statistics
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# Get the standard deviation of the values in the y-axis.
stdev = statistics.stdev(y_axis)
stdev
```

```python
plt.errorbar(x_axis, y_axis, yerr=stdev, capsize=3)
```

```python
fig, ax = plt.subplots()
ax.errorbar(x_axis, y_axis, yerr=stdev, capsize=3)
plt.show()
```

```python
plt.bar(x_axis, y_axis, yerr=stdev, capsize=3)
```

```python
fig, ax = plt.subplots()
ax.bar(x_axis, y_axis, yerr=stdev, capsize=3)
plt.show()
```

```python
import numpy as np
plt.barh(x_axis, y_axis)
plt.xticks(np.arange(0, 51, step=5.0))
plt.gca().invert_yaxis()
```

```python
fig, ax = plt.subplots()
ax.barh(x_axis, y_axis)
ax.set_xticks(np.arange(0, 51, step=5.0))
plt.show()
```

```python
from matplotlib.ticker import MultipleLocator
# Increase the size of the plot figure.
fig, ax = plt.subplots(figsize=(8, 8))
ax.barh(x_axis, y_axis)
ax.set_xticks(np.arange(0, 51, step=5.0))

# Create minor ticks at an increment of 1.
ax.xaxis.set_minor_locator(MultipleLocator(1))
plt.show()
```
